```python
from nltk.corpus import names
all_names = set(names.words())
from nltk.corpus import stopwords
stopwords = set(stopwords.words())
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
#1. LDA 토픽모델링
#series list로 만들기
df1_question = df1['question'].to_list()
df2_question = df2['question'].to_list()
df3_question = df3['question'].to_list()
df4_question = df4['question'].to_list()
df5_question = df5['question'].to_list()
#question 리스트만 딕셔너리로 저장
question_dic = {'df1':df1_question, 'df2':df2_question, 'df3':df3_question, 'df4':df4_question, 'df5':df5_question}
```

```python
def make_word(doc): #토픽모델링을 위한 전처리 함수
  cleaned_doc = []
  for word in doc.split(' '):
    word = word.lower()
    if word.isalpha() and word not in all_names and word not in stopwords and len(word) > 2: #알파벳이고 고유명사가 아니며 불용어가 아니고 2단어 이상인 경우
      cleaned_doc.append(lemmatizer.lemmatize(word))
  return cleaned_doc

word_dic = {}
for df in ['df1','df2','df3','df4','df5']:
  temps = [make_word(doc) for doc in question_dic[df]]
  word_dic[df] = temps
```

```python
import matplotlib.pyplot as plt
from gensim.models import CoherenceModel
from gensim.models.ldamodel import LdaModel
from gensim.corpora import Dictionary

for df in ['df1','df2','df3','df4','df5']:
  docs = word_dic[df]
  dictionary = Dictionary(docs)
  corpus = [dictionary.doc2bow(doc) for doc in docs]
  # 최적의 토픽 개수 찾기 5-15개까지 perplexity, coherence 계산
  start_topic_count = 5
  end_topic_count = 15
  perplexity_scores = []
  coherence_scores = []
  for i in range(start_topic_count, end_topic_count+1):
    model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=i, passes=10)
    perplexity = model.log_perplexity(corpus)
    perplexity_scores.append(perplexity)
    coherence_model_lda = CoherenceModel(model=model, texts=docs, dictionary=dictionary, coherence='c_v')
    coherence = coherence_model_lda.get_coherence()
    coherence_scores.append(coherence)
    print("토픽 개수: {}, Perplexity: {}, Coherence Score: {}".format(i, perplexity, coherence))

  # 결과를 시각화
  x = range(start_topic_count, end_topic_count+1)
  fig, ax1 = plt.subplots()
  ax1.plot(x, perplexity_scores, 'b-')
  ax1.set_xlabel('Topic Counts')
  ax1.set_ylabel('Perplexity', color='b')
  ax2 = ax1.twinx()
  ax2.plot(x, coherence_scores, 'r-')
  ax2.set_ylabel('Coherence Score', color='r')
  plt.show()
  #최적의 토픽 개수를 알기 위해 각 토픽 개수당 perplexity와 coherence score을 구하기 -> 최적의 토픽 개수 결정
```

```python
def preprocess_func(sentence):
    sentence = re.sub(r"([?.!,])", r" \1 ", sentence) #필요없는 특수기호 제거
    sentence = sentence.strip()
    sentence = sentence.strip()
    return sentence
```

```python
def convert_dataset(input_data, text, label):
    list_of_example = [data.Example.fromlist(row.tolist(), fields=[('text', text), ('label', label)])  for _, row in input_data.iterrows()]
    dataset = data.Dataset(examples=list_of_example, fields=[('text', text), ('label', label)])
    return dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(X, test_size = 0.3) #test size 0.3으로 설정하여 data split
```

```python
TEXT = data.Field(sequential=True, use_vocab=True, tokenize=tokenizer.morphs, lower=False, batch_first=True, fix_length=50) #데이터 최대 길이 50으로 설정
LABEL = data.LabelField(dtype = torch.float) #라벨 필드 생성
BATCH_SIZE = 64 #배치 사이즈 64로 설정

print(X_train.shape)
print(X_test.shape)

train_data = X_train.dropna() #말뭉치에서 nan 값을 제거함
test_data = X_test.dropna()


train_data['text'] = train_data['text'].apply(preprocess_func) #전처리 함수 적용하여 text에 저장
test_data['text']  = test_data['text'].apply(preprocess_func)

# split data
train_data, valid_data = train_test_split(train_data, test_size=0.3, random_state=32) #validation set 0.3으로 설정

print(len(train_data))
print(len(valid_data))
print(len(test_data))

print(train_data.shape)
print(valid_data.shape)  
print(test_data.shape)

train_data = convert_dataset(train_data, TEXT, LABEL)
valid_data = convert_dataset(valid_data, TEXT, LABEL)
test_data  = convert_dataset(test_data, TEXT, LABEL)


print(f'Number of training examples   : {len(train_data)}')
print(f'Number of validation examples : {len(valid_data)}')
print(f'Number of testing examples    : {len(test_data)}')


MAX_VOCAB_SIZE = 20000 #단어 갯수 20000개 저장

# 단어 사전 생성
TEXT.build_vocab(train_data, max_size = MAX_VOCAB_SIZE)
LABEL.build_vocab(train_data)

print(f"Unique tokens in TEXT vocabulary : {len(TEXT.vocab)}")
print(f"Unique tokens in LABEL vocabulary: {len(LABEL.vocab)}")

train_loader, valid_loader, test_loader = data.Iterator.splits( #batch size 개수대로 각각 train, test, valid dataset 나누기
  (train_data, valid_data, test_data),
  batch_size = BATCH_SIZE,
  sort = False,
  device = device)

print('Number of minibatch for training dataset   : {}'.format(len(train_loader)))
print('Number of minibatch for validation dataset : {}'.format(len(valid_loader)))
print('Number of minibatch for testing dataset    : {}'.format(len(test_loader)))
```

```python
class RNNModel(nn.Module):
    
    def __init__(
        self,
        input_size,
        word_vector_size,
        hidden_size,
        n_classes,
        n_layers=4,
        dropout_p=0.4
    ):
        #각 파라미터 값 default값으로 설정
        self.input_size = input_size
        self.word_vector_size = word_vector_size
        self.hidden_size = hidden_size
        self.n_classes = n_classes
        self.n_layers = n_layers
        self.dropout_p = dropout_p
        
        super().__init__()
        
        self.emb = nn.Embedding(input_size, word_vector_size) #단어 임베딩 실행
        self.rnn = nn.LSTM( #lstm model layer 쌓기
            input_size=self.word_vector_size,
            hidden_size=self.hidden_size,
            num_layers=self.n_layers,
            dropout=self.dropout_p,
            batch_first=True,
            bidirectional=True
        )
        
        self.linear_layer = nn.Linear(self.hidden_size*2, self.n_classes)
        self.softmax_layer = nn.LogSoftmax(dim=-1) #활성화함수 softmax 정의
        
    def forward(self, x):
        # |x| = (batch_size, length)
        x = self.emb(x)
        # |x| = (batch_size, length, word_vector_size)
        x, _ = self.rnn(x)
        # |x| = (batch_size, length, hidden_size*2)
        x = self.linear_layer(x[:,-1])
        # |x| = (batch_size, n_classes)
        y = self.softmax_layer(x)
        # |x| = (batch_size, n_classes)
        
        return y
```

```python
import matplotlib.pyplot as plt
class Trainer():

  def __init__(self, model, optimizer, crit):
      self.model = model
      self.optimizer = optimizer
      self.crit = crit

      super().__init__()

  def _train(self, train_loader, config):
      self.model.train()

      total_loss = 0

      for i, (x_i, y_i) in enumerate(train_loader): #각 train set batch 횟수만큼 반복
          y_hat_i = self.model(x_i)
          loss_i = self.crit(y_hat_i, y_i.squeeze().long())

          # Initialize the gradients of the model.
          self.optimizer.zero_grad()
          loss_i.backward() #역전파 실행

          self.optimizer.step() #옵티마이저 실행
          
          if config['verbose'] >= 2:
              print("Train Iteration(%d/%d): loss=%.4e" % (i + 1, len(train_loader), float(loss_i))) #loss값 출력

          # Don't forget to detach to prevent memory leak.
          total_loss += float(loss_i) #loss값 더하기

      return total_loss / len(train_loader) #최종 1 epoch당 loss값 출력

  def _validate(self, valid_loader, config):
      # Turn evaluation mode on.
      self.model.eval() #평가 모드 전환

      # Turn on the no_grad mode to make more efficintly.
      with torch.no_grad():
          total_loss = 0 #total loss 0으로 초기화

          for i, (x_i, y_i) in enumerate(valid_loader): #valid loader set만큼 반복
              y_hat_i = self.model(x_i)
              loss_i = self.crit(y_hat_i, y_i.squeeze().long())
              
              if config['verbose'] >= 2:
                  print("Valid Iteration(%d/%d): loss=%.4e" % (i + 1, len(valid_loader), float(loss_i))) #loss 출력

              total_loss += float(loss_i) #total loss에 loss 더하기

          return total_loss / len(valid_loader)

  def train(self, train_loader, valid_loader, config):
      lowest_loss = np.inf
      best_model = None
      train_loss_list=[]
      valid_loss_list=[]
      for epoch_index in range(config['n_epochs']): #epoch 횟수만큼 반복
          train_loss = self._train(train_loader, config) #train loss 모두 실행
          valid_loss = self._validate(valid_loader, config) #최종 valid loss 저장

          # You must use deep copy to take a snapshot of current best weights.
          if valid_loss <= lowest_loss: #valid loss가 이전에 저장한 가장 낮은 loss보다 낮을 경우
              lowest_loss = valid_loss #loss값 업데이트
              best_model = deepcopy(self.model.state_dict()) #best model 업데이트
          train_loss_list.append(train_loss)
          valid_loss_list.append(valid_loss)
          print("Epoch(%d/%d): train_loss=%.4e  valid_loss=%.4e  lowest_loss=%.4e" % ( #1 epoch 당 loss 값 출력
              epoch_index + 1,
              config['n_epochs'],
              train_loss,
              valid_loss,
              lowest_loss,
          ))
          
      # 그래프의 x축은 리스트의 인덱스로 설정
      x = range(len(train_loss_list))

      # 첫 번째 리스트에 대한 그래프 (파란색)
      plt.plot(x, train_loss_list, color='blue', label='train loss')

      # 두 번째 리스트에 대한 그래프 (빨간색)
      plt.plot(x, valid_loss_list, color='red', label='valid loss')

      # x축과 y축 라벨 설정
      plt.xlabel('epoch')
      plt.ylabel('loss Value')

      # 범례 추가
      plt.legend()

      # 그래프 보여주기
      plt.show()
      
      # Restore to best model.
      self.model.load_state_dict(best_model)
      return self.model
```

```python
from sklearn.metrics import classification_report, roc_auc_score, roc_curve
import numpy as np
```

```python
model_fn = "./model.pth"
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
def load(fn, device):
  d = torch.load(fn, map_location=device)
  
  return d['rnn'], d['config'], d['vocab'], d['classes']
```

```python
def test(model, test_loader):
  model.eval()
  
  correct_cnt, total_cnt = 0, 0
  y_true, y_pred, y_score=[], [], []
  with torch.no_grad():
    for x, y in test_loader:
      x, y = x.to(device), y.to(device)
      y_hat = model(x)
      _, predicted=torch.max(y_hat, dim=-1)
      correct_cnt += (y.squeeze() == torch.argmax(y_hat, dim=-1)).sum()
      total_cnt += len(x)

      y_true.extend(y.squeeze().tolist())
      y_pred.extend(predicted.tolist())
      y_score.extend(y_hat[:,1].tolist())

    accuracy = correct_cnt / total_cnt
    print("Accuracy: %.4f" % accuracy)

    report=classification_report(y_true, y_pred, target_names=["class 0", "class 1"], digits=4) #classification report
    print("Classification report:\n", report)

    roc_auc=roc_auc_score(y_true, y_score) #ROC AUC Score
    print("ROC-AUC Score: %.4f"%roc_auc)
    
    fpr, tpr, thresholds=roc_curve(y_true, y_score) # ROC Curve 그래프
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.show()
    
```
